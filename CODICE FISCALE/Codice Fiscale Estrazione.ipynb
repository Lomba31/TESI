```python
def CodFisc(tax_code):
    try: 
        if codicefiscale.is_valid(tax_code):
            d = codicefiscale.decode(tax_code)
            if d["birthplace"] != None:
                return d["birthplace"]["name"], d["birthplace"]["province"]
            else:
                return "Not Available", "Not Available"
        else:
            return "Not Valid", "Not Valid"
    except:
        return "Not Available", "Not Available"
citta, provincia = CodFisc("SBRTRN83L04C111V")
citta, provincia
```
