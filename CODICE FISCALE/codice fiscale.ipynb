```python
import pandas as pd
```

```python
df_comuni = pd.read_csv('listacomuni.txt', encoding="latin-1", sep=";", keep_default_na=False)
df_comuni
```

```python
df_codice_fiscale = pd.read_excel("codici_fiscali_770_2021.xlsx")
df_codice_fiscale
```

```python
df_codice_fiscale["CodFisco"] = df_codice_fiscale["tax_code"].str[-5:-1]
df_codice_fiscale
```

```python
df = pd.merge(df_codice_fiscale, df_comuni, how = "left", on = ["CodFisco"])
df
```

```python
df = df[["tax_code", "Comune", "Provincia"]]
df
```

```python
df["Comune"].fillna("Estero", inplace = True)
df["Provincia"].fillna("EE", inplace = True)
```

```python
df
```

```python
df.to_excel("COD AGGIORNATO.xlsx", index = False)
df.to_csv("COD AGGIORNATO.csv")
```
